# Buscando as contações de uma Ação

- **Ticker** é o código de uma ação.

```python
import yfinance
```

```python
codigo = input("Digite o código da ação: ")

dados = yfinance.Ticker(codigo).history("6mo")
```

```python
fechamento = dados.Close
```

```python
fechamento.plot()
```

# Gerar análises

- 6 meses
- cotação máxima
- cotação mínima
- cotação atual

```python
maxima = round(fechamento.max(),2)
minima = round(fechamento.min(),2)
atual = round(fechamento[-1],2)
```

```python
print(maxima)
print(minima)
print(atual)
```

# Enviar e-mail automaticamente

- abrir uma nova aba (CTRL + T)
- digitar o endereço do Gmail
- clicar enter
- clicar no botão Escrever
- preencher o destinatário (clicar tab)
- preencher assunto do e-mail (clicar tab)
- preencher corpo do e-mail
- clicar no botão enviar

```python
import pyautogui
import pyperclip
```

```python
# configurando o pyautogui para dar uma pausa de 3 segundos entre os passos
pyautogui.PAUSE = 3

# abrir uma nova aba (ctrl + t)
pyautogui.hotkey("ctrl", "t")

#digitar o endereço do Gmail
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

#clicar no botão escrever
pyautogui.click(x=1364, y=227)

#preencher o destinatário (clicar tab)
pyperclip.copy("jonathanaraujo92@icloud.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# preencher assunto (clicar tab)
pyperclip.copy("Análises diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# preencher corpo do e-mail
mensagem = f"""
Prezado Gestor,

Seguem as analises diárias da ação {codigo} referentes aos últimos seis meses:

Cotação máxima: R${maxima}
Cotação mínima: R${minima}
Cotação atual: R${atual}

Qualquer dúvida, estou à disposição!
"""

pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

pyautogui.click(x=1939, y=1000)

```
